```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('images/multi_faces.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray, cmap='gray')
```

```python
# load in casscade classifier
face_casscade = cv2.CascadeClassifier('detect_architect/haarcascade_frontalface_default.xml')
#run the detector
faces = face_casscade.detectMultiScale(gray, 4, 6)
```

```python
#print the detection found
print("We found {} faces in this image".format(str(len(faces))))
print("Their Coordinate and length/Width are as following")
print(faces)
```

```python
img_with_detection = np.copy(image)
# plot the corres detection boxes
# last argument is the width of the rectangle
for (x,y,w,h) in faces:
    cv2.rectangle(img_with_detection, (x,y), (x+w, y+h), (255,0,0),5)
    
plt.figure(figsize=(20,10))
plt.imshow(img_with_detection)
```
