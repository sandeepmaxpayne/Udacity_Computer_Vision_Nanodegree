```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('images/round_farms.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# gray and blur
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

gray_blur = cv2.GaussianBlur(gray, (3, 3), 0)

plt.imshow(gray_blur, cmap='gray')
```

```python
# for drawing circles on

circle_im = np.copy(image)
# Use Hough Circles to detect circles
circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, 1,
                          minDist=20,
                          param1=70,
                          param2=11,
                          minRadius=5,
                          maxRadius=20)
# convert circles to expected type
circles = np.uint16(np.around(circles))
#draw each one
for i in circles[0,:]:
    cv2.circle(circle_im, (i[0], i[1]), i[2], (0,255,0), 2)
    
    cv2.circle(circle_im, (i[0], i[1]), 2, (0,0,255), 3)

plt.imshow(circle_im)
print('cirlce shape: ', circles.shape)
```
