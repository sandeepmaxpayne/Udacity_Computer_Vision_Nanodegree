```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np

%matplotlib inline

#Read the image
image = cv2.imread('images/curved_lane.jpg')

plt.imshow(image)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
#Create a custom kernal
#using Sobel Operator
# Applying sobel filter means taking derivative of the image in x and y coordinates

# create a custom kernal

# 3 x 3 array for edge detection
sobel_y = np.array([[-1, -2, -1],
                   [ 0, 0, 0],
                   [ 1, 2, 1]])

#Apply sobel_y filter
filteres_image = cv2.filter2D(gray, -1, sobel_y)
#plt.imshow(filteres_image, cmap='gray')

#TODO Apply the sobel_x filter

sobel_x = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

filter_image_x = cv2.filter2D(gray, -1, sobel_x)

# to display the images

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("Filter using Sobel_X")
ax1.imshow(filter_image_x, cmap='gray')

ax2.set_title("Filter using Sobel_Y")
ax2.imshow(filteres_image, cmap='gray')

```
