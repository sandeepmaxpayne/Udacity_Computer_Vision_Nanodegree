```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
## TODO: Check out the images directory to see other images you can work with
# And select one!
image = cv2.imread('images/flamingos.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Prepare data for k-mean

#Reshape the  image into a 2D array of pixels and 3 color values RGB
pixel_val = image.reshape(-1, 3)

#convert to float type
pixel_val = np.float32(pixel_val)

```

Implementing K mean Clustering

```python
#define chopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)

# select value of k

k = 6

retval, labels, centers = cv2.kmeans(pixel_val, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
# center data into 8 bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

#rechape the data into original image dimensions
segmented_image = segmented_data.reshape((image.shape))
labels_reshape = labels.reshape(image.shape[0], image.shape[1])

plt.imshow(segmented_image)
```

```python
# Visualize the image
plt.imshow(labels_reshape==2, cmap='gray')
```

```python
# mask the image
cluster = 0
masked_image = np.copy(image)

#turn the mask green
masked_image[labels_reshape == cluster] = [0,0,250]

plt.imshow(masked_image)
```
