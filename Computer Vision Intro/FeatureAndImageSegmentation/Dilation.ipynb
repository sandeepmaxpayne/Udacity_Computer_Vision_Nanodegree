```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image = cv2.imread('images/j.png')
plt.imshow(image)
```

```python
# create a 5 x 5 kernal
kernal = np.ones((5,5), np.uint8)
```

```python
dilation = cv2.dilate(image, kernal, iterations=1)
```

```python
plt.imshow(dilation)
```

```python
# Erosion
erode = cv2.erode(image, kernal, iterations=1)
plt.imshow(erode)
```

```python
copy_img = np.copy(image)
gray = cv2.cvtColor(copy_img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
```

# Opening to remove noise outside the object in a  region (Erosion => Dilation)

```python
opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernal)
plt.imshow(opening)
```

# Closing => To remove the holes present within an object(dilation -> Erosion )

```python
closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernal)
plt.imshow(closing)
```
