```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib qt
```

```python
#read the image
image = mpimg.imread("image/car_green_screen.jpg")
print(image.shape)
```

```python
image.size
```

```python
plt.imshow(image)
```

```python
cp_image = np.copy(image)
plt.imshow(cp_image)
```

```python
# color threshold

lower_green = np.array([0, 180, 0])
upper_green = np.array([250, 255, 250])
```

```python
# create mask
mask = cv2.inRange(cp_image, lower_green, upper_green)

plt.imshow(mask, cmap='gray')
```

```python
# mask the image to the car
copy_image = np.copy(cp_image)
copy_image[mask != 0] = [0, 0, 0]
plt.imshow(copy_image)
```

```python
#mask and add background
back_img = mpimg.imread("image/sky.jpg")
plt.imshow(back_img)
back_img.shape
```

```python

back_img = cv2.cvtColor(back_img, cv2.COLOR_RGB2BGR)
crop_back = back_img[0:450, 0:660]
crop_back[mask==0] = [0, 0, 0]

plt.imshow(crop_back)
```

```python
# create complete image
complete_img = copy_image + crop_back
plt.imshow(complete_img)
```
