```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
image = cv2.imread("image/water_balloons.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Plot the color channels
r = image[:, :, 0]
g = image[:, :, 1]
b = image[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title("Red Axis")
ax1.imshow(r, cmap="gray")

ax2.set_title('Green Axis')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue Axis')
ax3.imshow(b, cmap='gray')
```

```python
# convert from RGB to HSB
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

#plot hsv axis
h = hsv[:,:,0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title("Hue")
ax1.imshow(h, cmap='gray')

ax2.set_title("Saturation")
ax2.imshow(s, cmap="gray")

ax3.set_title("Value")
ax3.imshow(v, cmap="gray")
```

# Defining hue and pink selction thresholds

```python
lower_hue = np.array([160,0,0])
higher_hue = np.array([180, 255, 255])

lower_pink = np.array([180, 0, 100])
higher_pink = np.array([255, 255, 230])
```

```python
#mask the image
mask_range = cv2.inRange(image, lower_pink, higher_pink)
mask_image = np.copy(image)

mask_image[mask_range == 0] = [0, 0, 0] 

plt.imshow(mask_image)
```

```python
#mask the image for hsv
hasv_range = cv2.inRange(hsv, lower_hue, higher_hue)
mash_hue = np.copy(image)

mash_hue[hasv_range == 0] = [0, 0, 0]
plt.imshow(mash_hue)
```
